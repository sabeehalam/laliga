```
df['Points']=pd.to_numeric(df['Points'],errors='coerce')
df.sort_values(by='Points',ascending=False)
df.head()
```

```
def Goal_diff_count(GoalsFor, GoalsAgainst ):
  Goal_diff=GoalsFor-GoalsAgainst
  return Goal_diff
```

```
df['GoalsFor']=pd.to_numeric(df['GoalsFor'],errors='coerce')
df['GoalsAgainst']=pd.to_numeric(df['GoalsAgainst'],errors='coerce')
df.head()
```

```
df['Goal_diff']=df.apply(lambda x: Goal_diff_count(x['GoalsFor'],x['GoalsAgainst']),axis=1)
df.head()
```

```
df[df['Goal_diff']==df['Goal_diff'].max()][['Team','Goal_diff']]
```

```
df[df['Goal_diff']==df['Goal_diff'].min()][['Team','Goal_diff']]
```

```
df['GamesWon']=pd.to_numeric(df['GamesWon'])
df['GamesPlayed']=pd.to_numeric(df['GamesPlayed'])
df['Win_percent']=(df['GamesWon']/df['GamesPlayed'])*100
df.head()
```

```
df['Win_percent']=df['Win_percent'].fillna(value=0)
```

```
df.sort_values(by='Win_percent',ascending=False)
```
